### SQLite database for NPPES created on 2/23.

```python
import pandas as pd
import sqlite3
from tqdm.notebook import tqdm
```

```python
db = sqlite3.connect('../data/nppes_lite.sqlite')
```

```python
for chunk in tqdm(pd.read_csv('../data/NPPES_Data_Dissemination_February_2021/npidata_pfile_20050523-20210207.csv',
                              chunksize = 10000)):
    chunk.columns = [x.lower().replace(' ', '_') for x in chunk.columns]      # Clean up the column names
    chunk.to_sql('nppes', db, if_exists = 'append', index = False)            # Append the chunk to a calls table
```

```python
db.close()
```
